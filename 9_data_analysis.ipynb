### Python Data Analysis

Python is widely used to analyze data for business and science purposes. This typically includes describing, summarizing, and representing data visually to identify trends and patterns.

Python's popularity stems from robust features to acquire & manage data, as well as a built-in `statistics` module and numerous third-party libraries.


### Descriptive Statistcs

Pythonâ€™s built-in `statistics` library supports basic descriptive statistics and is good in situations where one's datasets are not too large or you don't wnat to rely on third-party libraries.

The module is not intended as a competitor to full-featured libraries such as NumPy, SciPy, but is sufficient for common descriptive statistics.

````python
from statistics import mean, median, mode, stdev, variance, quantiles
data = [1, 2, 3, 4, 4, 6]
print(f"mean {mean(data):.2f}")      # compute average value of data array
print("median", median(data))  # return middle value of the data array
print("mode", mode(data))      # return most common data point in the data array

# compute measures of data variability
print(f"std deviation {stdev(data):.2f}")
print(f"variance {variance(data):.2f}")

# Divide data into n continuous intervals with equal probability. n=4 by default
print("quantiles", quantiles(data))
````

### Visualization with matplotlib

`matplotlib` is a python library supporting a wide range of data plots.

You may need to install matplotlib on your computer like so:

```
python -m pip install -U matplotlib
```

matplotlib includes several key sub-libraries, such as `pyplot` which accepts Python lists or tuples with numeric values and can generate a variety of plot types.

`pyplot` can calculate default axis ranges and tick marks.

### Basic line plot

````python
# Import the necessary packages
import matplotlib.pyplot as plt

x_values = range(5) # values for x-axis
y_values = [3,12,7,9,18] # values for y-axis

# create sub-plot showing the data as a line
plt.plot(x_values, y_values, label='my data')

# optional text elements
plt.title('Student grades')
plt.ylabel('# of students')
plt.xlabel('Grade')
plt.legend()

# Show the plot
plt.show()
````

### Scatter plot
The data above could also be displayed as a scatter plot like so:

````python
plt.scatter(x_values,y_values, label='my data')
plt.show()
````

### Bar chart
or as a bar chart or horizontal bar chart:

````python
plt.bar(x_values,y_values, label='my data', align="center") # show vertical bars
plt.show()
plt.barh(x_values,y_values, label='my data') # show horizontal bars
plt.show()
````

### Pie charts

pyplot can display data as a pie chart to show relative proportions:

````python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
````

### Subplots
A plot can include several sub-plots of the same or different types:

````python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26])
plt.show()
````
